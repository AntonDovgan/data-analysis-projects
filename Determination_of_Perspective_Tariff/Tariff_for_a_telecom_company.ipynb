### Сводные таблицы

Найдем для каждого пользователя:

* количество сделанных звонков и израсходованных минут разговора по месяцам;
* количество отправленных сообщений по месяцам;
* объем израсходованного интернет-трафика по месяцам.

#### Сводная таблица <code>data_calls_pivot</code>

```python
calls_temp = data_calls.merge(data_users, on='user_id', how='inner') # объеденеям с data_users
data_calls_pivot = calls_temp.pivot_table(index = ['user_id', 'month', 'tariff', 'age', 'city'], values = 'duration', aggfunc=['count','sum'])
data_calls_pivot.columns  = ['calls_count', 'calls_duration']
data_calls_pivot['calls_duration'] = data_calls_pivot['calls_duration'].astype('int')
data_calls_pivot
# data_calls_pivot.info()
```

#### Сводная таблица <code>data_intenet_pivot</code>

```python
internet_temp = data_internet.merge(data_users, on = 'user_id', how = 'inner') # объеденеям с data_users
data_internet_pivot = internet_temp.pivot_table(index = ['user_id', 'month', 'tariff', 'age', 'city'], values = 'mb_used', aggfunc = ['count', 'sum'])
data_internet_pivot.columns = ['session_count', 'mb_sum']

# Округляем мегабайты в Гигабайты по правилам компании
data_internet_pivot['gb'] = data_internet_pivot['mb_sum']/1024
data_internet_pivot['gb'] = data_internet_pivot['gb'].apply(np.ceil)
data_internet_pivot['gb'] = pd.to_numeric(data_internet_pivot['gb'], downcast='integer') # float => int
data_internet_pivot
```

#### Сводная таблица <code>data_msg_pivot</code>

```python
msg_temp = data_msg.merge(data_users, on = 'user_id', how = 'inner')
data_msg_pivot = msg_temp.pivot_table(index = ['user_id', 'month', 'tariff', 'age', 'city'], values = 'id', aggfunc = ['count'])
data_msg_pivot.columns = ['msg_count']
data_msg_pivot.reset_index()
```

### Объединение сводных таблиц calls, internet, messages

```python
dfs = [data_calls_pivot, data_internet_pivot, data_msg_pivot]
data_main = reduce(lambda left,right: pd.merge(left,right,on= ['user_id','month', 'tariff', 'city'], how='outer'), dfs).reset_index()
data_main .drop('mb_sum', axis = 1, inplace = True) # удаляем столбец mb_sum
# пропуски заполняем нолями
data_main['calls_count'] = data_main['calls_count'].fillna(0)
data_main['calls_duration'] = data_main['calls_duration'].fillna(0)
data_main['session_count'] = data_main['session_count'].fillna(0)
data_main['gb'] = data_main['gb'].fillna(0)
data_main['msg_count'] = data_main['msg_count'].fillna(0)
# переводим в формат целого числа
data_main['calls_count'] = data_main['calls_count'].astype('int')
data_main['calls_duration'] = data_main['calls_duration'].astype('int')
data_main['session_count'] = data_main['session_count'].astype('int')
data_main['gb'] = data_main['gb'].astype('int')
data_main['msg_count'] = data_main['msg_count'].astype('int')

data_main.columns = 'user_id', 'month', 'tariff', 'city','calls_count', 'duration', 'session_count', 'gb_used', 'msg_count'
data_main.info()
```

### Посесячная выручка по каждому пользователю.

```python
data_tariff['gb_per_month_included'] = data_tariff['mb_per_month_included']/1024
data_tariff
```

```python
def revenue_func(row):
    tariff = row['tariff']
    duration = row['duration']
    gb_used = row['gb_used']
    msg = row['msg_count']    
    
    if tariff == 'smart':
        extra_duration = duration - data_tariff.loc[0, 'minutes_included']
        extra_gb = gb_used - data_tariff.loc[0, 'gb_per_month_included']
        extra_msg = msg - data_tariff.loc[0, 'messages_included']
        
        if extra_duration < 0: extra_duration = 0
        if extra_gb < 0: extra_gb = 0
        if extra_msg < 0: extra_msg = 0
            
        return  (extra_duration * data_tariff.loc[0, 'rub_monthly_fee'] +
                 extra_gb * data_tariff.loc[0, 'rub_per_gb'] + 
                extra_msg * data_tariff.loc[0, 'rub_per_message'] + 
                data_tariff.loc[0, 'rub_monthly_fee'])
    else:
        extra_duration = duration - data_tariff.loc[1, 'minutes_included']
        extra_gb = gb_used - data_tariff.loc[1, 'gb_per_month_included']
        extra_msg = msg - data_tariff.loc[1, 'gb_per_month_included']
        
        if extra_duration < 0 : extra_duration = 0
        if extra_gb < 0: extra_duration = 0
        if extra_msg < 0: extra_msg = 0
    
        return (extra_duration * data_tariff.loc[1, 'minutes_included'] + 
                extra_gb * data_tariff.loc[1, 'gb_per_month_included'] +
                extra_msg * data_tariff.loc[1, 'messages_included'] +
                data_tariff.loc[1,'rub_monthly_fee'])
        
data_main['revenue'] = data_main.apply(revenue_func, axis = 1)
data_main.head(10)
```

```python
data = data_main.pivot_table(index = ['user_id','month', 'tariff'], values = ('duration', 'calls_count', 'session_count', 'gb_used' , 'msg_count', 'revenue'))
data
```

### Проверка гипотез
#### Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.

H0  - cредние выручи пользователей тарифов «Ultra» и «Smart» равны  
H1  - cредние выручки пользователей тарифов «Ultra» и «Smart» различаются  
α = 0.05%

```python
smart = data_main.query('tariff == "smart"')['revenue']
ultra = data_main.query('tariff == "ultra"')['revenue']

alpha = 0.05
results = st.ttest_ind(smart, ultra)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### Cредняя выручка пользователи из Москвы отличается от выручки пользователей из других регионов
H0 - средняя выручка пользователей из Москвы равна среденей выручке пользователей из других городов  
H1 - средняя выручка пользователей из Москвы отличается от средней выручки пользоватлей из других городов
α = 0.05%

```python
moscow  = data_main.query('city == "Москва"')['revenue']
another  = data_main.query('city != "Москва"')['revenue']

alpha = 0.05
results = st.ttest_ind(moscow, another)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
